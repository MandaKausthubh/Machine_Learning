```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SizeOfSample = 10**6

c1 = np.zeros(SizeOfSample, dtype=int)
c2 = np.zeros(SizeOfSample, dtype=int)
c3 = np.zeros(SizeOfSample, dtype=int)

for i in range(SizeOfSample):
    Experiment = np.random.randint(low=0, high=11, size=1000)
    c1[i] = Experiment[0]
    c2[i] = Experiment[np.random.randint(len(Experiment))]
    c3[i] = min(Experiment)

print(type(c1[0]))
print(type(c2[0]))
```

```python
plt.hist(c1, bins=list(range(0,12)))
plt.show()
```

```python
plt.hist(c2, bins=list(range(0,12)))
plt.show()
```

```python
plt.hist(c3, bins=list(range(0,12)))
plt.show()
```
